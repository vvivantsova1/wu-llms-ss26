```python
#Error message
import os
import anthropic
from dotenv import load_dotenv

# 1. Load the environment variables from the .env file
load_dotenv()

# 2. Initialize the Anthropic client
# It will automatically find the ANTHROPIC_API_KEY in your environment
try:
    client = anthropic.Anthropic()
except anthropic.AnthropicError as e:
    print("Error: Could not initialize client. Check your .env file.")
    raise e

# 3. Define the test case
test_instruction = "Analysiere den folgenden steuerrechtlichen Sachverhalt und nenne die relevanten österreichischen Paragraphen."
test_input = "Ein selbstständiger IT-Berater kauft einen neuen Laptop für 2.500 Euro, den er zu 80% beruflich und zu 20% privat nutzt. Er möchte wissen, wie er diesen im Rahmen der Einkommensteuererklärung als Betriebsausgabe absetzen kann und ob eine AfA notwendig ist."

formatted_prompt = f"""Instruction: {test_instruction}

Input (Sachverhalt):
{test_input}

Bitte nenne nur die relevanten österreichischen Paragraphen (z.B. 'EStG 1988 §4 Abs1')."""

print("Sending request to Claude 3 Haiku...")

# 4. Call the API
response = client.messages.create(
    model="claude-3-haiku-20240307",
    max_tokens=256,
    temperature=0.1, 
    system="Du bist ein präziser und hochqualifizierter KI-Assistent für das österreichische Steuerrecht (EStG). Deine Aufgabe ist es, rechtliche Sachverhalte zu analysieren und ausschließlich die exakten Gesetzesstellen zu nennen.",
    messages=[
        {"role": "user", "content": formatted_prompt}
    ]
)

# 5. Print the output
print("\n--- Claude 3 Haiku Output ---")
print(response.content[0].text)
```
